```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print(">>> ADIM 1: Veri Seti Okunuyor...")
# Arkadaşının kodunda değişken ismini tamamen değiştiriyoruz
covid_tablosu = pd.read_csv('Covid Data.csv') # Dosya adın farklıysa burayı güncelleyebilirsin

print(f"Toplam Kayıt (Hasta): {covid_tablosu.shape[0]}")
print(f"Sütun (Özellik) Sayısı: {covid_tablosu.shape[1]}\n")

print(">>> ADIM 2: Eksik Veri Analizi ve Temizliği...")
# Eksik değerleri bulalım
bos_degerler = covid_tablosu.isnull().sum()
print("Eksik veri barındıran sütunlar:")
print(bos_degerler[bos_degerler > 0])

# Temizlik işlemi (Yine dropna ama farklı bir değişkene atıyoruz)
temiz_veri = covid_tablosu.dropna()
print(f"\nTemizleme tamamlandı! Kalan hasta sayısı: {temiz_veri.shape[0]}\n")

print(">>> ADIM 3: Veri Görselleştirme...")
plt.figure(figsize=(9, 5))
# Senin kodundaki 'viridis' yerine 'magma' renk paletini kullanıyoruz ki grafik tamamen farklı görünsün
sns.heatmap(covid_tablosu.isnull(), cbar=False, cmap='magma')
plt.title('Eksik Verilerin Dağılım Haritası (Temizlik Öncesi)')
plt.show()

print("\n-> BAŞARILI: Veri temizlendi, makine öğrenmesi aşamasına geçilebilir.")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

print("--- Modeller İçin Veri Hazırlığı Başlıyor ---")
ml_verisi = temiz_veri.copy()

ml_verisi['TEST_SONUCU'] = ml_verisi['CLASIFFICATION_FINAL'].apply(lambda x: 1 if x <= 3 else 0)

ozellikler = ml_verisi.drop(columns=['CLASIFFICATION_FINAL', 'DATE_DIED', 'TEST_SONUCU'])
hedef = ml_verisi['TEST_SONUCU']

# HATA ÇÖZÜMÜ BURADA: Eğer 45 binden az veri kaldıysa hata verme, olanı kullan
alinacak_sayi = min(45000, len(ozellikler))
ozellikler = ozellikler.sample(n=alinacak_sayi, random_state=101)
hedef = hedef.loc[ozellikler.index]

x_egitim, x_test, y_egitim, y_test = train_test_split(ozellikler, hedef, test_size=0.25, random_state=101)

print(f"\nSeçilen Hasta Sayısı: {alinacak_sayi}")
print("--- Modeller Eğitiliyor ve Test Ediliyor (Biraz zaman alabilir) ---")

algoritmalar = {
    "Lojistik Regresyon": LogisticRegression(max_iter=1500),
    "Karar Ağacı": DecisionTreeClassifier(random_state=101),
    "Rastgele Orman": RandomForestClassifier(n_estimators=150, random_state=101),
    "K-En Yakın Komşu (KNN)": KNeighborsClassifier(n_neighbors=7)
}

for isim, model in algoritmalar.items():
    print(f"\n>>>> Model: {isim} <<<<")

    model.fit(x_egitim, y_egitim)
    tahminler = model.predict(x_test)

    dogruluk = accuracy_score(y_test, tahminler)
    kesinlik = precision_score(y_test, tahminler, zero_division=0)
    duyarlilik = recall_score(y_test, tahminler, zero_division=0)

    print(f"Accuracy (Doğruluk) Oranı  : % {dogruluk*100:.2f}")
    print(f"Precision (Kesinlik) Oranı : % {kesinlik*100:.2f}")
    print(f"Recall (Duyarlılık) Oranı  : % {duyarlilik*100:.2f}")

    matris = confusion_matrix(y_test, tahminler)
    plt.figure(figsize=(3, 2))
    sns.heatmap(matris, annot=True, fmt='d', cmap='Blues')
    plt.title(f'{isim} Matrisi')
    plt.ylabel('Gerçek Değerler')
    plt.xlabel('Tahmin Edilenler')
    plt.show()

print("\n-> İŞLEM TAMAM: 4 Algoritma da başarıyla test edildi! 🚀")
```
